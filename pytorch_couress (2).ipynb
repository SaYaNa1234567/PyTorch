```python
import numpy as np
from scipy.integrate import quad
from scipy.stats import norm

mu = 0.0
sigma = 0.1
a = -0.1
b = 0.1

def normal_pdf(x):
    coefficient = 1 / (sigma * np.sqrt(2 * np.pi))
    exponent = np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return coefficient * exponent

integral_result, error = quad(normal_pdf, a, b)
```

simple data, i cooked it

```python
# Create *known* parameters
weight = 0.7
bias = 0.3

# Create data
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
y = weight * X + bias

X[:10], y[:10]
```

```python
y[:5]
```

```python
# Create train/test split
train_split = int(0.8 * len(X)) # 80% of data used for training set, 20% for testing
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]

len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
y_train
```

```python
class Linear_regression(nn.Module):
  def __init__(self):
    super().__init__()
    self.wight=nn.Parameter(torch.randn(1))
    self.bias=nn.Parameter(torch.randn(1))
  def forward(self,x:torch.tensor):
    return self.wight*x+self.bias
```

```python
model_0=Linear_regression()
```

```python
loss=nn.L1Loss()
optimation=torch.optim.SGD(params=model_0.parameters(),lr=0.001)
```

```python
epoch=20
li1=[]


for i in range(epoch):
  model_0.train()
  y_pred=model_0(X_train)

  l1_loss=loss(y_pred,y_train)
  li1.append(l1_loss)
  optimation.zero_grad() ## we need to set it to zero because we don't wanna accumlate each gradiant of each layer to each other
  l1_loss.backward() ## backward
  optimation.step() ## to update each parm in each layer of arch
  ## test
  model_0.eval()
test_process=model_0(X_test)
print(loss(y_pred,y_train))
print(test_process)
```
